**From Lecture Note 8, Exercise 1** A process known to be in statistical control has xbar = 20 and s = 1.2. Specifications set for the output is to be within 16 and 24 only.

(a) Estimate the process capability with an appropriate process capability ratio.


Assuming there is no known target for process centering, we can evaluate the C_p ratio as $C_p = \frac{24 - 16}{6 \times 1.2} = 1.11$

However, if $\overline{\overline{x}}$ is indicative of the process center, then we can use the C_{pk} ratio $C_{pk} = \min\{ \frac{24 - 20}{3 \times 1.2 },  \frac{20 - 16}{3 \times 1.2 } \} = \min\{ 1.11,  1.11 \} = 1.11$

There is no need to evaluate for the $C_{pm}$ measure. Note that $T = \frac{1}{2}(24 + 16) = 20 = \overline{\overline{x}}$. Thus, $V = 0$ and $\hat{C}_{pm} = \hat{C}_{pk}$

**From Lecture Note 8, Exercise 4** The data in the following table presents the cycle time (in hours) to process small loan applications. Prepare a normal probability plot of these data. The loan agency has a promised decision time to potential customers of 24 hours. Based on the data in the table and the normal probability plot, what proportion of the customers will experience longer waiting times?

```r
loans <- c(
    16.3, 16.3, 19.3, 15.1, 22.2,
    19.1, 18.5, 18.3, 18.7, 20.2,
    22.0, 14.7, 18.0, 18.9, 19.1,
    10.6, 18.1, 19.6, 20.8, 16.5,
    19.3, 14.6, 17.8, 15.6, 22.5,
    17.6, 17.2, 20.9, 14.8, 18.2,
    16.4, 18.2, 19.4, 14.1, 16.4,
    19.6, 17.5, 17.1, 21.7, 20.8
)
```

The `e1071` library provides an easy-to-use interface for creating probability plots.

```r
probplot(loans)
```

```r
loans.qcc <- qcc(loans, type = "xbar.one")
```

Let's assume a number of parameters for this process. We can create a qw

```r
process.capability(loans.qcc, spec.limits = c(9,27), target = 18.05, std.dev = 2.7, 
                   confidence.level = 0.95)
```
